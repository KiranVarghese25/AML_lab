```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
```

```python
#loading the dataset
df=pd.read_csv("/content/Program_4_Dataset.csv")
df.head()
```

```python
#Dropping the unwanted columns
df=df.drop(['index','OS_Time_Stamp','Humidity','TimeStamp'],axis=1)
```

```python
#checking for null values
df.isnull().sum()
```

```python
#setting the value of X and Y
X=df[['Temperature']]
Y=df[['Pressure']]
```

```python
#model 1{ degree 2 and test size 0.44}

#splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.44, random_state=42)
p=PolynomialFeatures(degree=2, include_bias=True)
X_train_P=p.fit_transform(X_train)
X_test_P=p.fit_transform(X_test)

#building the model
model = LinearRegression()
model.fit(X_train_P, Y_train)
y_pred1=model.predict(X_test_P)

#Evaluating the model
r2 = r2_score(Y_test, y_pred1)
mse = mean_squared_error(Y_test, y_pred1)
mae = mean_absolute_error(Y_test, y_pred1)

print("R^2 Score: {0}, MSE: {1}, MAE: {2}".format(r2,mse,mae))

#Graph
plt.scatter(X_test, Y_test, color='blue', label='True values')
plt.scatter(X_test, y_pred1, color='orange', label='Predicted values')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.show()
```
